# Upsampling

```python
from pandas import read_csv
from pandas import datetime

series = read_csv('TP_month.csv', header=0, parse_dates=[0], index_col=0, squeeze=True)
upsampled = series.asfreq('D')
print(upsampled.head(32))
```

```python
import matplotlib.pyplot as plt
series = read_csv('TP_month.csv', header=0, parse_dates=[0], index_col=0, squeeze=True)
upsampled = series.resample('D')
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
interpolated.plot()
plt.show()
interpolated.to_csv('TP_month_day.csv')
```

# Downsampling

```python
series = read_csv('TP.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
resample = series.resample('M')
month = resample.mean()
print(month.head(32))
month.plot()
plt.show()
month.to_csv('TP_month.csv')
```

```python
resample = series.resample('Q')
quarterly_mean_sales = resample.mean()
print(quarterly_mean_sales.head(32))
quarterly_mean_sales.plot()
plt.show()
```
