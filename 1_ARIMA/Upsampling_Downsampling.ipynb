# Upsampling

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('TP_month.csv')
dataset['date'] = pd.to_datetime(dataset['date'])
dataset = dataset.set_index('date')
upsampled = dataset.resample('W').mean()
print(upsampled.head(32))
```

```python
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
interpolated.plot()
plt.show()
interpolated.to_csv('month_week.csv')
```

# Downsampling

```python
series = read_csv('TP.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
resample = series.resample('M')
month = resample.mean()
print(month.head(32))
month.plot()
plt.show()
month.to_csv('TP_month.csv')
```

```python
resample = series.resample('Q')
quarterly_mean_sales = resample.mean()
print(quarterly_mean_sales.head(32))
quarterly_mean_sales.plot()
plt.show()
```
